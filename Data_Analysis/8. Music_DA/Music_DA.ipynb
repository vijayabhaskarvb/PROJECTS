```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('MusicData.csv')
dataset = df.copy()
dataset.head()
```

```python
dataset.info()
dataset.describe()
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
dataset = dataset.drop(['Value (Actual)'], axis = 1)
dataset.isna().sum().sum()
```

```python
dataset.duplicated().sum()
```

```python
dataset.head(2)
```

```python
dataset['Year'].unique()
```

```python
sns.set(style= 'dark')
```

```python
dataset['Year'].value_counts().sort_values(ascending=False).head(20).plot(kind='barh', color = ['Blue', 'Orange'], figsize = (6,4))
plt.title('Year Distribution')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation = 0)
plt.show()
```

```python
dataset['Metric'].value_counts().sort_values(ascending=False).plot(kind='bar', color = ['Blue', 'Orange'], figsize = (6,4))
plt.title('Metric Distribution')
plt.xlabel('Metrics')
plt.ylabel('Count')
plt.xticks(rotation = 0)
plt.show()
```

```python
dataset['Format'].value_counts().sort_values(ascending=False).head(20).plot(kind='pie', figsize = (6,4))
plt.title('Top 20 Format Distribution')
plt.xlabel('Metrics')
plt.ylabel('')
plt.show()
```

```python
dataset.groupby('Year')['Metric'].value_counts().sort_values(ascending = False).head(20).plot(kind = 'bar', color = ['Blue', 'Orange'], figsize = (6,4))
plt.title('Year & Metric Distribution')
plt.xlabel('Year & Metric')
plt.ylabel('')
plt.xticks(rotation = 90)
plt.show()
```

```python
dataset.groupby('Format')['Metric'].value_counts().sort_values(ascending = False).head(20).plot(kind = 'bar', color = ['Blue', 'Orange'], figsize = (6,4))
plt.title('Format & Metric Distribution')
plt.xlabel('Format & Metric')
plt.ylabel('')
plt.xticks(rotation = 90)
plt.show()
```
