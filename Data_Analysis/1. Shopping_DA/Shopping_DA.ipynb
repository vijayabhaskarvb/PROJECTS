```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('shopping_trends.csv')
dataset = df.copy()
dataset.head()
```

```python
dataset.info()
dataset.describe()
```

```python
dataset = dataset.drop(['Customer ID'], axis = 1)
dataset.head(2)
```

```python
dataset['Age'].mean()
```

```python
sns.set(style='dark')
```

```python
dataset['Gender'].value_counts().plot(kind = 'bar', figsize=(6,4), color = ['Blue', 'Orange'])
plt.title("Gender Distribution")
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
dataset['Item Purchased'].value_counts().sort_values(ascending=False).head(10).plot(kind = 'bar', figsize=(6,4), color = ['Blue', 'Orange'])
plt.title('Top 10 Item Purchased')
plt.xlabel('Items')
plt.ylabel('Count')
plt.show()
```

```python
dataset['Category'].value_counts().sort_values(ascending=False).plot(kind = 'pie', figsize=(6,4))
plt.title('Category Distribution')
plt.xlabel('Category')
plt.ylabel('')
plt.show()
```

```python
dataset['Size'].value_counts().sort_values(ascending=False).plot(kind = 'pie', figsize=(6,4))
plt.title('Size Distribution')
plt.xlabel('Sizes')
plt.ylabel('')
plt.show()
```

```python
dataset['Preferred Payment Method'].value_counts().plot(kind = 'barh', figsize=(6,4), color = ['Blue', 'Orange'])
plt.title('Preferred Payment Methods')
plt.xlabel('Count')
plt.ylabel('Payment Method')
plt.show()
```

```python
dataset['Frequency of Purchases'].value_counts().sort_values(ascending=True).plot(kind = 'barh', figsize=(6,4), color = ['Blue', 'Orange'])
plt.title('Frequency of Purchase')
plt.xlabel('Count')
plt.ylabel('Frequency')
plt.show()
```

```python
dataset.head(2)
```

```python
dataset['Review Rating'].mean()
```

```python
dataset.groupby('Category')['Review Rating'].mean().sort_values(ascending=False).plot(kind = 'bar', figsize=(6,4), color = ['Blue', 'Orange'])
plt.title('Average Rating by Category')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.show()
```

```python
dataset.groupby('Gender')['Purchase Amount (USD)'].sum().plot(kind = 'bar', figsize=(6,4), color = ['Blue', 'Orange'])
plt.title('Sum of Purchase Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Purchase Amount')
plt.show()
```

```python
dataset.groupby('Category')['Purchase Amount (USD)'].sum().plot(kind = 'pie', figsize=(6,4))
plt.title('Sum of Purchase Amount by Category')
plt.xlabel('Category')
plt.ylabel('')
plt.show()
```
