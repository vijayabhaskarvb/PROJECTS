```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Supplement_Sales_Weekly_Expanded.csv')
dataset = df.copy()
dataset.head()
```

```python
dataset['Product Name'].value_counts().sort_values(ascending = True).plot(kind = 'bar', color = ['Blue', 'Orange'], figsize = (6,4))
plt.title('Products Distribution')
plt.xlabel('Products')
plt.ylabel('Count')
plt.xticks(rotation = 45)
plt.show()
```

```python
dataset['Category'].value_counts().sort_values(ascending = False).plot(kind = 'bar', color = ['Blue', 'Orange'], figsize = (6,4))
plt.title('Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation = 45)
plt.show()
```

```python
dataset['Location'].value_counts().sort_values(ascending = False).plot(kind = 'pie', figsize = (6,4))
plt.title('Location Distribution')
plt.xlabel('Location')
plt.ylabel('')
plt.show()
```

```python
dataset['Platform'].value_counts().sort_values(ascending = True).plot(kind = 'barh', color = ['Blue', 'Orange'], figsize = (6,4))
plt.title('Platform Distribution')
plt.xlabel('Count')
plt.ylabel('Platform')
plt.show()
```

```python
dataset.groupby('Category')['Revenue'].sum().sort_values(ascending = False)
```

```python
dataset.groupby('Category')['Units Sold'].sum().sort_values(ascending = False)
```

```python
dataset.groupby('Category')['Revenue'].sum().sort_values(ascending = False).plot(kind = 'bar', color = ['Blue', 'Orange'], figsize = (6,4))
plt.title('Total Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Total Revenue')
plt.xticks(rotation = 45)
plt.show()
```

```python
dataset.groupby('Location')['Units Sold'].sum().sort_values(ascending = False).plot(kind = 'barh', color = ['Blue', 'Orange'], figsize = (6,4))
plt.title('Total Units Sold by Location')
plt.xlabel('')
plt.ylabel('Location')
plt.show()
```

```python
dataset.groupby('Location')['Platform'].value_counts().sort_values(ascending = False).plot(kind = 'bar', color = ['Blue', 'Orange'], figsize = (6,4))
plt.title('Platforms used by Location')
plt.xlabel('Platform & Location')
plt.ylabel('')
plt.show()
```
