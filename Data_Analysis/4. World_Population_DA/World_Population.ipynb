```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('world_population_data.csv')
dataset = df.copy()
dataset.head()
```

```python
dataset.info()
dataset.describe()
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
dataset.duplicated().sum()
```

```python
sns.set(style='dark')
```

```python
dataset.groupby('country')['2023 population'].sum().sort_values(ascending=False).head(10).plot(kind = 'bar', figsize=(6,4), color = ['Blue', 'Orange', 'Yellow', 'Green', 'Red', 'Teal', 'Indigo'])
plt.title('Country by Population in 2023')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```

```python
dataset.groupby('country')['2022 population'].sum().sort_values(ascending=True).head(10).plot(kind = 'barh', figsize=(6,4), color = ['Blue', 'Orange', 'Yellow', 'Green', 'Red', 'Teal', 'Indigo'])
plt.title('Country by Population')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show()
```

```python
dataset.groupby('country')['2020 population'].sum().sort_values(ascending=False).head(10).plot(kind = 'pie', figsize=(6,4))
plt.title('Country by Population in 2020')
plt.xlabel('Country')
plt.ylabel('')
plt.show()
```

```python
total_population = dataset['2023 population'].sum()
print(f"Total Population in the world in 2023: {total_population}")
```
